```python
from rembg import remove
from PIL import Image
import os
```

```python
input_root = "."
output_root = 'Masked'
```

```python
subfolders = [
    'training_data/ripe_lemons',
    'training_data/unripe_lemons',
    'test_data/ripe_lemons',
    'test_data/unripe_lemons'
]
```

```python
for folder in subfolders:
    input_path = os.path.join(input_root, folder)
    output_path = os.path.join(output_root, folder)

    os.makedirs(output_path, exist_ok=True)  # Create output directory

    for filename in os.listdir(input_path):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            try:
                input_img_path = os.path.join(input_path, filename)
                output_img_path = os.path.join(output_path, filename)

                with Image.open(input_img_path) as img:
                    img_no_bg = remove(img)
                    output_img_path = os.path.splitext(output_img_path)[0] + '.png'
                    img_no_bg.save(output_img_path, format='PNG')

                print(f"✅ Processed: {input_img_path} → {output_img_path}")

            except Exception as e:
                print(f"⚠️ Error processing {filename}: {e}")
```
